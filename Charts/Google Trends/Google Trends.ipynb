```python
import pandas as pd
import re
import plotly.graph_objects as go
from IPython.display import display, clear_output, Markdown

layout = go.Layout(
  margin=go.layout.Margin(
        l=0,
        r=0,
        b=0,
        t=0,
    ),
    legend=dict(
    orientation="h",
    y=1.02,
    xanchor="right",
    yanchor="bottom",
    x=1,
    font=dict(
        size=22,
     ),
)
)
```

```python
df = pd.read_csv('Daten/Google Trends geringe Relevanz 01.01.09 bis 26.05.21 weltweit.csv', header=1)
df = df.replace('<1', 0)
df = df.rename(columns=lambda x: re.sub(r'^([^:]+):.*$', r"\1", x))

cols = df.columns.drop('Monat') 
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

fig = go.Figure(layout=layout)

fig.add_trace(go.Scatter(x=df.iloc[:,0], y=df.iloc[:,1], name=df.columns[1], line=dict(color='black', width=2)))
fig.add_trace(go.Scatter(x=df.iloc[:,0], y=df.iloc[:,2], name=df.columns[2], line=dict(color='#6cadf6', width=2)))
fig.add_trace(go.Scatter(x=df.iloc[:,0], y=df.iloc[:,3], name=df.columns[3], line=dict(color='#3b57f6', width=2)))
fig.add_trace(go.Scatter(x=df.iloc[:,0], y=df.iloc[:,4], name=df.columns[4], line=dict(color='#aa1b31', width=2)))
fig.add_trace(go.Scatter(x=df.iloc[:,0], y=df.iloc[:,5], name=df.columns[5], line=dict(color='#90c43e', width=2)))

fig.update_layout(yaxis_title='Relatives Suchinteresse')

fig.show()
fig.write_image("Suchinteresse geringe Relevanz.pdf", width=800, height=300)
```
